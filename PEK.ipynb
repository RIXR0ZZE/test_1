```python
import numpy as np
import matplotlib.pyplot as plt

#Создаем случайные координаты точек
r = np.sqrt(np.random.uniform(0, 1, 5000))
theta = np.random.uniform(0, 2*np.pi, 5000)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.scatter(x, y, s = 4)
plt.show()
```

```python
#Считаем расстояние от точки до центра окружности по т.Пифагора
def pifagor(a,b):
    c = np.sqrt(a ** 2 + b ** 2)
    return c
```

```python
# Считаем расстояние каждой точки до центра
# ПЫ: Чем больше точек, тем ближе среднее значение радиуса будет к 0,(6), т.е. 2/3. Через интегралы лень выводить, с вуза помню.
R = []
for i in range(0, len(x)):
    R.append(pifagor(x[i], y[i]))
```

```python
sum(R)/len(R)
```
