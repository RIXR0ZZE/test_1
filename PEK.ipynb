```python
import numpy as np
import matplotlib.pyplot as plt

#   
r = np.sqrt(np.random.uniform(0, 1, 5000))
theta = np.random.uniform(0, 2*np.pi, 5000)
x = r * np.cos(theta)
y = r * np.sin(theta)
plt.scatter(x, y, s = 4)
plt.show()
```

```python
#        .
def pifagor(a,b):
    c = np.sqrt(a ** 2 + b ** 2)
    return c
```

```python
#      
# :   ,        0,(6), .. 2/3.    ,   .
R = []
for i in range(0, len(x)):
    R.append(pifagor(x[i], y[i]))
```
